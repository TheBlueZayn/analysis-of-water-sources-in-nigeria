# Analysis Of Source Of Drinking Water In Nigeria-2018

## Analysis of Data
- Import Libraries
- Pandas for data analysis and wrangling 
- Matplotlib and Seaborn for Visualisation


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data
- Load CSV dataset into a DataFrame with Pandas

```python
df = pd.read_csv("cfafrica-_-data-team-_-outbreak-_-covid19-_-openafrica-uploads-_-nigeria-source-of-drinking-wate.csv")
```

## Explore Data
- .head() to view the first five rows of dataframe
- .info() to get insights on the entire dataset
- .corr() to view the corellation between columns with numbers
- Build a heatmap to better visualise the correlation matrix

```python
df.head()
```

```python
df.info()
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, cmap="Blues")
plt.title("Correlation Matrix Between The Sources Of Drinking Water In Nigeria")
plt.savefig("corrmatrix", dpi=300, bbox_inches="tight"), ;
```

# Subset Data


## Subset Data into Urban and Rural
- Use pandas to subset
- Visualize Subset with matplotlib barplot
- Customize plot size, labels and title

```python
urban_rural = df.loc[1:2, :]
urban_rural
```

```python
urban_rural.set_index("Geography").plot(kind="bar",
                                        figsize=(20,6)) # Specifies a barplot with set width & height in inches
plt.xlabel("Geographical Area")
plt.ylabel("Count")
plt.title("Sources Of Drinking Water In Nigeria", fontdict={"fontsize":20})
plt.savefig('rural_urban.png', dpi=300, bbox_inches='tight')
```

## Subset into Geo-Political Zones
- Use Pandas to subset data
- Visualise a scatterplot to show the relationship between improved water and water piped in the 6 geopolitical zones in Nigeria


```python
geo_zones = df.loc[3:8, :]
geo_zones
```

```python
sns.scatterplot(x="Improved water source", y="Water piped into dwelling", data=geo_zones, hue="Geography")
plt.title("Improved Water Source as a factor of Water Pumped Into Dwelling", fontdict={'fontsize':10, 'fontweight':20})
plt.savefig('relationships', dpi=300, bbox_inches='tight')

;
```

## Subset Into States
- Subset with pandas
- Visualizing a horizontal barplot of the best 15 states with improved water sources 
- Visualizing a horizontal barplot of the worst 15 states with improved water sources

```python
state =  df.loc[9:].set_index("Geography")
state.shape
```

```python
state
```

```python
best_states = state["Improved water source"].sort_values(ascending=False)[:15].sort_values()

best_states.plot(kind="barh",color="purple")
plt.xlabel("Count")
plt.title("States In Nigeria With The Best Water Sources")
plt.savefig('beststates', dpi=300, bbox_inches='tight');



```

```python
worst_states = state["Unimproved water source"].sort_values(ascending=False)[:15].sort_values()
worst_states.plot(kind="barh", color="#1fa6c4")
plt.xlabel("Count")
plt.title("States In Nigeria With The Worst Water Sources")
plt.savefig('worststates', dpi=300, bbox_inches='tight');

```

```python
state["Improved water source"].max()
highest_state = df[df["Improved water source"] == 89.5]
highest_state.reset_index(drop=True).plot(kind="bar", figsize=(12,6))
plt.xlabel("Abia State")
plt.ylabel("Count")
plt.title("State With Highest Improved Water Source Count")
plt.savefig('higheststate', dpi=300, bbox_inches='tight');
```

## Subset Into Lagos State
- Subset data
- Visualise the various sources of drinking water in Lagos state

```python
state.loc["Lagos"].sort_values(ascending=False).sort_values().plot(kind="barh",
                                                                   color="k",
                                                                   figsize=(10,10))
plt.xlabel("Count")
plt.ylabel("Water Source")
plt.title("Sources Of Water In Lagos State")
plt.savefig("lagos", dpi=300, bbox_inches="tight");
```
