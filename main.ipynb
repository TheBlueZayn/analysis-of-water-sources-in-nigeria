```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Data
- Load CSV dataset into a DataFrame with Pandas

```python
df = pd.read_csv("cfafrica-_-data-team-_-outbreak-_-covid19-_-openafrica-uploads-_-nigeria-source-of-drinking-wate.csv")
```

## Explore Data

```python
print("The DataFrame Shape is", df.shape)
```

```python
df.info()
```

```python
corr = df.corr()
corr
```

```python
sns.heatmap(corr, cmap="Blues")
plt.title("Correlation Matrix Between the Sources of Drinking Water in Nigeria")
plt.savefig("corrmatrix", dpi=300, bbox_inches="tight"), ;
```

# Subset Data


## Subset Data into Urban and Rural
- Use pandas to subset
- Visualize Subset with matplotlib barplot
- Customize plot size, labels and title

```python
urban_rural = df.loc[1:2, :]
urban_rural
```

```python
urban_rural.set_index("Geography").plot(kind="bar",
                                        figsize=(20,10)) 
plt.xlabel("Geographical Areas", fontdict={"fontsize":15})
plt.xticks(rotation=0, fontsize=15)
plt.ylabel("Count", fontdict={"fontsize":15})
plt.title("Sources of Drinking Water in Nigeria", fontdict={"fontsize":20})
plt.savefig('rural_urban.png', dpi=300, bbox_inches="tight")
```

## Subset Into Geo-Political Zones
- Use pandas to subset data
- Visualise a scatterplot to show the relationship between improved water and water piped in 


```python
geo_zones = df.loc[3:8, :]
geo_zones
```

```python
sns.scatterplot(x="Improved water source", y="Water piped into dwelling", data=geo_zones, hue="Geography")
plt.title("Improved Water Source Against Water Pumped into Dwelling", fontdict={'fontsize':10, 'fontweight':20})
plt.savefig('relationships', dpi=300, bbox_inches='tight');
```

## Subset Into States
- Subset with pandas
- Visualisea horizontal barplot of the best 15 states with improved water sources 
- Visualise a horizontal barplot of the worst 15 states with improved water sources
- Visualise a barplot of the best state and worst state
- Visuaiise the various water sources in Lagos State

```python
states =  df.loc[9:].set_index("Geography")
states.shape
```

```python
states
```

```python
best_states = states["Improved water source"].sort_values(ascending=False)[:15].sort_values()

best_states.plot(kind="barh",color="purple")
plt.xlabel("Count")
plt.title("States in Nigeria with the Best Water Sources")
plt.savefig('beststates', dpi=300, bbox_inches='tight');



```

```python
worst_states = states["Unimproved water source"].sort_values(ascending=False)[:15].sort_values()
worst_states.plot(kind="barh", color="#1fa6c4")
plt.xlabel("Count")
plt.title("States in Nigeria with the Worst Water Sources")
plt.savefig('worststates', dpi=300, bbox_inches='tight');

```
